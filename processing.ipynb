```python
# %pip install pandas
import pandas as pd
```

```python
# Import the csv and visualize the first rows
df = pd.read_csv('scraping.csv')
df.head()
```

```python
# Add id and the price in euros
df['id'] = df.index
df['Price (euros)'] = round(df['Price (pounds)'] * 1.17, 2)   # conversion value on 06/12/2023
df.head()
```

```python
# %pip install selenium
from selenium import webdriver

# Connect to the text generator API
url = 'https://deepai.org/chat/text-generator'
driver = webdriver.Chrome()
driver.get(url)
```

```python
# Initiatlize an empty list to save the generated texts
texts = []

# Iterate rows
for index, book in df.iterrows():
    # Find the last chatbox
    chatboxes = driver.find_elements('class name','chatbox')
    chatbox= chatboxes[-1]
    # Request the text of the book
    chatbox.send_keys('The text of  '+book['Title']+' is: \n')

    # Wait until the response is fully generated
    while (len(chatboxes) == len(driver.find_elements('class name','chatbox'))):
        pass

    # Find the last outputbox
    outputboxes = driver.find_elements('class name', 'outputBox')
    outputbox = outputboxes[-1]
    # Add the generated text to the list
    texts.append(outputbox.text.split('\n')[:-4][0])
```

```python
texts
```

```python
# %pip install deep_translator
from deep_translator import GoogleTranslator

# Translate the texts into Spanish and Chinese
texts_es = GoogleTranslator(source='en', target='es').translate_batch(texts)
texts_ch = GoogleTranslator(source='en', target='zh-CN').translate_batch(texts)

df2 = df[:len(texts)] # This is only done because I couldn't get the text of all the books

# Add the text and translated texts to the dataframe
df2['Text'] = texts
df2['Text (es)'] = texts_es
df2['Text (ch)'] = texts_ch
df2.head()
```

```python
# Save csv
df2.to_csv('out.csv')
```
